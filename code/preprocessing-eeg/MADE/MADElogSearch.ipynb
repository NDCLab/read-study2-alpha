```python
# Code to identify participants processed in the .out file from a slurm job (as it's running)
# Convienent for debugging live script issues when a job is running
# Search .out file for "MADE pipeline completed for" and print matching lines

filepath = "/home/data/NDClab/analyses/read-study2-alpha/code/preprocessing-eeg/MADE/READEEGMADEpreprocess-2698631.out"
processedPattern = "MADE pipeline completed"
totalProcessed = 0

with open(filepath, "r", errors="replace") as f:
    for i, line in enumerate(f, 1):
        if processedPattern in line:
            print(f"Line {i}: {line.rstrip()}")
            totalProcessed += 1
print("Total Processed =", totalProcessed)

# Identify participants with no usable data
failedPattern = "No usable data"

with open(filepath, "r", errors="replace") as f:
    for i, line in enumerate(f, 1):
        if failedPattern in line:
            print(f"Line {i}: {line.rstrip()}")
```
