```python
# Excercise
# Pythagorean theorem: a**2 + b**2 = c**2
# Get two numbers from the user for the length of side a and the length of side b
# Compute and print the length of side c
import math as mt
# Initialize variables
sidea = input('What is side A? ')
sideb = input('What is side B? ')
# Calculate side C
sidec_squared = float(sidea)**2 + float(sideb)**2
sidec = mt.sqrt(sidec_squared)
# Give output
print('Side C is: ' + str(sidec))
```

```python
# Tuples
# To create a tuple, use parenthesis instead of brackets
atuple = ( 3,'4',5 )
print(atuple) # Appears similar to a list
atuple*3 # CAn do same multiplying with tuple as with lists
# What's the difference between tuples and lists?
# Tuple can be immutable and a list is mutable
alist = [ 3,'4',3 ]
atuple = ( 3,'4',3 )
# With a list an item can be replaced in the list. In a tuple this is not possible
print(atuple)
print(alist)
alist[0] # indexing first item of a list
alist[0] = 5
print(alist) # changed the first number of the list
#atuple[0] = 5 # errors
type(list(atuple))
```
